```python
# greyscale
imgGrey = cv.imread('E:\OCV\Elon Musk.jpeg',cv.IMREAD_GRAYSCALE)
imgGrey = cv.resize(imgGrey,(450,500))  # to resize the video 

cv.imshow('image',imgGrey)
cv.waitKey()
cv.destroyAllWindows()
```

```python
# BW
(thresh,imgbw) = cv.threshold(imgGrey,0,255,cv.THRESH_BINARY | cv.THRESH_OTSU)
cv.imshow('image',imgbw)
cv.waitKey()
cv.destroyAllWindows()
```

```python
# edges
edges = cv.Canny(imgGrey,150,150)
cv.imshow('image',edges)
cv.waitKey(1000)
cv.destroyAllWindows()
```

```python
# Using matplotlib

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv.imread('E:\OCV\Elon Musk.jpeg')
plt.imshow(img)
```

```python
image = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
grey = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
plt.imshow(grey)
```

```python
# rectangle
img_rect = cv.rectangle(image,pt1=(100,400),pt2=(350,50),color=(50,200,255),thickness=2)
plt.imshow(img_rect)
```

```python
# Filled rectangle - give -ve thickness
img_rectfill = cv.rectangle(image,pt1=(100,400),pt2=(350,50),color=(50,200,255),thickness=-2)
plt.imshow(img_rectfill)
```

```python
# # circle
# img_cir = cv.circle(image,center=(200,110),radius=50,color=(100,200,255),thickness=2)

# plt.imshow(img_cir)
```

```python
image = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)

# circle
img_cir = cv.circle(image,center=(210,210),radius=150,color=(0,0,0),thickness=5)

plt.imshow(img_cir)
```

```python
image = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)

# Text
img_text = cv.putText(image,text='Elon',org=(250,1400),fontFace=cv.FONT_ITALIC,fontScale=10,color=(250,50,100),thickness=10)

plt.imshow(img_text)
```
