```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

## Face detection

```python
face_cascade = cv.CascadeClassifier('E:\OCV\haarcascade_frontalface_default.xml')
face_cascade
```

```python

img = cv.imread("E:\OCV\Elon Musk.jpeg")
plt.imshow(img)
```

```python
image = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Detecting face
faces = face_cascade.detectMultiScale(image,1.1,4,minSize = (30,30))
faces
```

```python
# no of faces
len(faces)
```

```python
# drawing rectangle on the detected face (through the detected coordinates)

for (x,y,w,h) in faces:
    cv.rectangle(image,(x-5,y-5),(x+w+5,y+h+5),(255,10,0),20)


plt.imshow(image)
```

### Face detection from Video

```python

# cap = cv.VideoCapture(f"Video_sample01.mp4")


cap=cv.VideoCapture(0)  #to get input from  webcam
```

```python
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Our operations on the frame come here
    grey = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(grey,1.1,4,minSize = (30,30))

    for (x,y,w,h) in faces:
        cv.rectangle(frame,(x,y),(x+w,y+h),(100,200,0),2)

    #frame2 = cv.resize(frame,(200,200))  # to resize the video 
    cv.imshow("video",frame)
    if cv.waitKey(1)& 0xFF ==27:
        break

cap.release()
cv.destroyAllWindows()
```

## Eye detection

```python
eye_cascade = cv.CascadeClassifier("haarcascade_eye.xml")

image = cv.imread('E:\OCV\Elon Musk.jpeg')

eye = eye_cascade.detectMultiScale(image,1.1 ,4 ,minSize=(30,30))
```

```python
eye
```

```python
len(eye)
```

```python
for (x,y,w,h) in eye:
    cv.rectangle(image,(x,y),(x+w,y+h),(255,100,255),20)

plt.imshow(image)
```

### Eye detection from webcam

```python
cap=cv.VideoCapture(0) 
```

```python
while True:
    ret,frame = cap.read()
    grey = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    eye = eye_cascade.detectMultiScale(grey,1.1 ,4 ,minSize=(30,30))
    for(x,y,w,h) in eye:
        cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    
    
    cv.imshow("video",frame)
    if cv.waitKey(1)& 0xFF ==27:
        break
cap.release()
cv.destroyAllWindows()
```
