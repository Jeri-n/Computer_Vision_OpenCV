```python
import mediapipe as mp
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img  = cv.imread("E:\OCV\Projects\Tracking\hands.jpg")
plt.imshow(img)
```

```python
plt.imshow(img[:,:,: :-1])   #shortcut
```

```python
# initialize the hand class and store it in a variable
mp_hands = mp.solutions.hands

# set the hand function which will hold the land mark points
hands = mp_hands.Hands(static_image_mode=True,max_num_hands=6,min_detection_confidence=0.3)

#set up the drawing function of hands landmark on the image
mp_drawing = mp.solutions.drawing_utils
```

```python
plt.title('Sample image')
plt.axis('off')
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
results = hands.process(cv.cvtColor(img,cv.COLOR_BGR2RGB))

if results.multi_hand_landmarks:
    for hand_no, hand_landmarks in enumerate(results.multi_hand_landmarks):
        mp_drawing.draw_landmarks(image=img,landmark_list=hand_landmarks,connections=mp_hands.HAND_CONNECTIONS)

    plt.title("Resultant image")
    plt.axis("off")
    plt.imshow(img[:,:,::-1])
    plt.show()
```
